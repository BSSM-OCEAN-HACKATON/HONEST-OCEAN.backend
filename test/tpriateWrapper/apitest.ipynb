```python
import httpx
import urllib
import pprint   

BASE_URL="https://pub-api.tpirates.com/v2/www/retail-price"
```

```python
def get_market_price(fish_name: str):
    POSTFIX="/price/aggregate/region?keyword={}&orderState=default&page=0&size=6"
    encoded_url = POSTFIX.format(urllib.parse.quote(fish_name))
    
    res = httpx.get(BASE_URL + encoded_url).json()

    # 가장 일치도가 높은 결과가 0번쨰에 오므로 0번 인덱스를 불러옴
    # 그리고 처음 나오는 avgprice가 kg 기준
    return res['content'][0]['avgPrice']


pprint.pprint(get_market_price("고등어"))
```
