```python
import httpx
import json

SERVICE_KEY = "ea3fc0775f925c0065b920574022dcfff1c9ed64e3a8545dc0b479a51b4430e4"

def fetch_and_search_fish(search_name: str):
    url = "http://apis.data.go.kr/6260000/KjMarketCost/getDailyCost"
    params = {
        "serviceKey": SERVICE_KEY,
        "pageNo": "1",
        "numOfRows": "200", 
        "resultType": "json",
        "fishName": search_name # Server-side filtering
    }
    
    print(f"Searching for '{search_name}' via API parameter...")
    
    try:
        response = httpx.get(url, params=params)
        
        if response.status_code != 200:
            print(f"❌ API Error: {response.status_code}")
            return
            
        data = response.json()
        items = data.get("response", {}).get("body", {}).get("items", {}).get("item", [])
        
        if not items:
             print(f"⚠️ No matches found for '{search_name}'.")
             return
        
        print(f"✅ Found {len(items)} matches for '{search_name}':")
        for item in items[:10]: # Show top 10
            price = item.get('avgCost') or item.get('maxCost') or 'N/A'
            print(f"- {item.get('fishName')} | {price}원 | {item.get('goodScale', 'N/A')}")
            
    except Exception as e:
        print(f"❌ Error: {e}")

# Example Usage
fetch_and_search_fish("연어")
```
